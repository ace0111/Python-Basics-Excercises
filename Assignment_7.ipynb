1. The re module is responsible for generating Regex objects in Python.
2. Raw strings are often used in Regex objects to avoid having to escape backslashes, as backslashes are commonly used in regular expressions to denote special characters.
3. The search() method returns a Match object if the pattern is found, or None if the pattern is not found.
4. You can use the group() method on a Match object to get the actual strings that match the pattern.
    
import re
# define the pattern to match
pattern = r'(\d\d)-(\d\d)-(\d\d\d\d)'

# create a Match object by searching the pattern in a string
match = re.search(pattern, 'My birthday is on 07-10-1990')

# get the actual strings that match the pattern
day = match.group(1)
month = match.group(2)
year = match.group(3)

# print the results
print(f"My birthday is on {day}/{month}/{year}")

6. To match real parentheses and periods in a regular expression, you need to escape them with a backslash.
7. The findall() method returns a string list if the regex pattern has no groups, or a list of string tuples if the pattern has one or more groups.
8. In standard expressions, the | character means "or".
9. In regular expressions, the dot character (.) stands for any single character except a newline character.
    An example of using the . character in a regular expression would be the pattern b.t, which would match any string containing the letters 'b' and 't' with any character in between them. For example, it would match "bat", "bet", "bit", "b*t", etc.

10.In regular expressions, the `+` and `*` characters are quantifiers used to indicate repetition of the preceding pattern. 

- `+` indicates that the preceding pattern must appear one or more times. For example, the pattern `ab+` will match the letter 'a' followed by one or more occurrences of the letter 'b' (e.g. "ab", "abb", "abbb", etc.).

- `*` indicates that the preceding pattern may appear zero or more times. For example, the pattern `ab*` will match the letter 'a' followed by zero or more occurrences of the letter 'b' (e.g. "a", "ab", "abb", "abbb", etc.).

In summary, `+` means "one or more" and `*` means "zero or more" of the preceding pattern.
11. {4} matches exactly four occurrences of the preceding character or group, while {4,5} matches between four and five occurrences of the preceding character or group.

    In regular expressions, the notation {m} specifies that the previous character or group should match exactly m times. For example, the regular expression "a{3}" will match the string "aaa", but not "aa" or "aaaa".

On the other hand, the notation {m,n} specifies that the previous character or group should match at least m times and at most n times. For example, the regular expression "a{2,4}" will match the strings "aa", "aaa", and "aaaa", but not "a" or "aaaaa".

Therefore, the difference between {4} and {4,5} is that the former will only match exactly four occurrences of the previous character or group, while the latter will match either four or five occurrences.

12. In regular expressions, the \d shorthand character class matches any digit character (0-9), the \w shorthand character class matches any word character (a-z, A-Z, 0-9, and underscore), and the \s shorthand character class matches any whitespace character (space, tab, newline, etc.).
13. In regular expressions, the \D shorthand character class matches any character that is not a digit, the \W shorthand character class matches any character that is not a word character, and the \S shorthand character class matches any character that is not a whitespace character.
14. .*? matches zero or more occurrences of any character (except a newline character) in a non-greedy way, while .* matches zero or more occurrences of any character (except a newline character) in a greedy way.
15. The syntax for matching both numbers and lowercase letters with a character class is [0-9a-z].
16. To make a normal expression in regex case-insensitive, you can pass the re.IGNORECASE flag as the second argument to re.compile().
17. The . character normally matches any character except a newline character. If re.DOTALL is passed as the second argument to re.compile(), the . character will also match newline characters.

19. Passing re.VERBOSE as the second argument to re.compile() allows you to use whitespace and comments within the regular expression pattern, making it more readable.
20. The regex to match a number with a comma for every three digits is r'^\d{1,3}(,\d{3})*$'.
21. The regex to match the full name of someone whose last name is Watanabe is r'[A-Z][a-z]*\sWatanabe'.
